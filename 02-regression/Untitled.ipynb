```python
class Employee:
    def _init_(self, employee_id,first_name,last_name,email,salary):
        self.employee_id = employee_id
        self.first_name = first_name
        self.last_name = last_name
        self.email = email
        self.salary = salary

    def get_employee_id(self):
        return self.employee_id

    def get_first_name(self):
        return self.first_name

    def get_last_name(self):
        return self.last_name

    def get_email(self):
        return self.email

    def get_salary(self):
        return self.salary

    def set_employee_id(self, employee_id):
        self.employee_id = employee_id

    def set_first_name(self, first_name):
        self.first_name = first_name

    def set_last_name(self, last_name):
        self.last_name = last_name

    def set_email(self, email):
        self.email = email

    def set_salary(self, salary):
        self.salary = salary


    def total_earning(self):
        pass

    def _str_(self):
        return f"Employee ID: {self.employee_id}, First Name: {self.first_name}, Last Name: {self.last_name}, Email: {self.email}, Salary: {self.salary}"


class Manager(Employee):

    def init(self, employee_id,first_name,last_name,email,salary, department, number_of_direct_report, management_support_allowance, rate):
       super().__init__(employee_id,first_name,last_name,email,salary)
       self.__department = department
       self.__number_of_direct_report = number_of_direct_report
       self.__management_support_allowance = management_support_allowance
       self.__rate = rate


    def get_department(self):
      return self.__department

    def set_department(self, department):
        self.__department = department


    def get_number_of_direct_report(self):
      return self.__number_of_direct_report


    def set_number_of_direct_report(self, number_of_direct_report):
      self.__number_of_direct_report = number_of_direct_report


    def get_management_support_allowance(self):
      return self.__management_support_allowance

    def set_management_support_allowance(self, management_support_allowance):
      self.__management_support_allowance = management_support_allowance

    def get_rate(self):
      return self.__rate

    def set_rate(self, rate):
      if rate > 0.6:
         raise ValueError("Rate must be less than 60 %")
      elif rate < 0:
         raise ValueError("Rate must be greater than 0%")
      else:
         self.__rate = rate

    def total_earnings(self):
      return self._salary + self.rate * self._salary



    def str(self):
        return f"Employee ID: {self.employee_id}, First Name: {self.first_name}, Last Name: {self.last_name}, Email: {self.email}, Salary: {self.salary}, Department: {self._department}, Number of Direct Report: {self.number_of_direct_report}, Management Support Allowance: {self.management_support_allowance}, Rate: {self._rate}"


class Director(Employee):

    def __init__(self, employee_id, first_name, last_name, email, salary, department, annual_bonus):
        super()._init_(employee_id, first_name, last_name, email, salary)
        self.__department = department
        self.__annual_bonus = annual_bonus


    def get_department(self):
        return self.__department

    def get_annual_bonus(self):
        return self.__annual_bonus

    def set_department(self, department):
        self.__department = department

    def set_annual_bonus(self, annual_bonus):
        self.__annual_bonus = annual_bonus

    def total_earning(self):
        return super().get_salary() + self.__annual_bonus



director = Director(1, "John", "Doe", "john@gmail.com", 100000, "IT", 50000)

manager  = Manager(20, "Yaw", "George", "george@gmail.com",2500, "IT", 3, 2500)

print(director.total_earning())

print(manager.total_earning())
```
